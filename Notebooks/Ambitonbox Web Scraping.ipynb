```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import time

company_names = []
ratings = []
reviews_count = []
salary_count = []
interview_count = []
jobs_count = []
company_sector = []
locations = []
benefits_count = []
page_num = []

base_url = "https://www.ambitionbox.com/list-of-companies?campaign=homepage_companies_widget"

headers = {
    'User-Agent': 'Mozilla/5.0',
    'Accept-Language': 'en-US,en;q=0.9'
}

total_time = time.time()

for i in range(1, 150):
    start_time = time.time()
    url = f"{base_url}&page={i}"

    page = requests.get(url, headers=headers)
    soup = BeautifulSoup(page.text, "html.parser")

    cards = soup.find_all('div', class_='companyCardWrapper')

    if not cards:
        print(f"No data found on page {i}, stopping...")
        break

    for card in cards:

        # Company Name
        name = card.find('h2', class_='companyCardWrapper__companyName')
        company_names.append(name.text.strip() if name else np.nan)

        # Rating
        rating = card.find('div', class_='rating_star_container')
        ratings.append(rating.text.strip() if rating else np.nan)

        # Actions (Reviews, Salaries, etc.)
        actions = card.find_all('span', class_='companyCardWrapper__ActionTitle')

        review = salary = interview = job = benefit = np.nan

        for act in actions:
            label = act.text.strip()
            value = act.find_previous_sibling('span')

            if label == "Reviews":
                review = value.text.strip() if value else np.nan
            elif label == "Salaries":
                salary = value.text.strip() if value else np.nan
            elif label == "Interviews":
                interview = value.text.strip() if value else np.nan
            elif label == "Jobs":
                job = value.text.strip() if value else np.nan
            elif label == "Benefits":
                benefit = value.text.strip() if value else np.nan

        reviews_count.append(review)
        salary_count.append(salary)
        interview_count.append(interview)
        jobs_count.append(job)
        benefits_count.append(benefit)

        # Sector & Location
        info = card.find('span', class_='companyCardWrapper__interLinking')
        if info:
            parts = info.text.strip().split('|')
            company_sector.append(parts[0].strip() if len(parts) > 0 else np.nan)
            locations.append(parts[1].strip() if len(parts) > 1 else np.nan)
        else:
            company_sector.append(np.nan)
            locations.append(np.nan)

       

    print(f"Page {i} done in {time.time() - start_time:.2f}s")

print("Total Time:", time.time() - total_time)
```

```python
import pandas as pd
Ambition_df = pd.DataFrame({"Company_names":company_names,
                            "Ratings":ratings,
                            "Reviews_count":reviews_count,
                            "Salary_count":salary_count,
                            "Interview_count":interview_count,
                            "Jobs_count":jobs_count,
                            "Company_sector":company_sector,
                            "Locations":locations,
                            "Benefits_count":benefits_count,
                            "Page_num":page_num})
Ambition_df.to_csv("Web ambitionbox dataset.csv",index = False)
```
